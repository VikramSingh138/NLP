```python
import re

def get_morphological_features(word, noun_set):
    if word.endswith("es"):
        stem = word[:-2]  # remove "es"
        if re.search(r'(s|x|z|ch|sh)$', stem):
            if stem in noun_set:
                return f"{stem}+N+PL"
            else:
                return "Invalid Word"
    if word.endswith("ies"):
        stem = word[:-3] + "y"
        if stem in noun_set:
            return f"{stem}+N+PL"
        else:
            return "Invalid Word"

    if word.endswith("s"):
        stem = word[:-1]
        if stem in noun_set and not word.endswith("ss"):
            return f"{stem}+N+PL"
        else:
            return "Invalid Word"
    if word in noun_set:
        return f"{word}+N+SG"

    return "Invalid Word"

def load_nouns(filename="brown_nouns.txt"):
    with open(filename, "r", encoding="utf-8") as f:
        nouns = {line.strip().lower() for line in f if line.strip()}
    return nouns

if __name__ == "__main__":
    noun_set = load_nouns()

    test_words = ["foxes", "fox", "tries", "try", "bags", "bag", "foxs", "watchees", "Dog", ""]

    for word in noun_set:
        result = get_morphological_features(word.lower(), noun_set)
        print(f"{word} → {result}")
```
