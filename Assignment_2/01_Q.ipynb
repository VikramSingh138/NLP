```python
def is_valid_simplified_english_word(word):
    state = "q0"
    
    for char in word:
        if state == "q0":
            if 'a' <= char <= 'z':
                state = "q1"
            else:
                return "Not Accepted"
        elif state == "q1":
            if 'a' <= char <= 'z':
                continue
            else:
                return "Not Accepted"
    
    return "Accepted" if state == "q1" else "Not Accepted"
```

```python
test_words = [
    "cat", "dog", "a", "zebra",         # Accepted
    "dog1", "1dog", "DogHouse",         # Not Accepted
    "Dog_house", " cats", "Cat", "dog!", # Not Accepted
    "", " ", "z"                        # "", " " rejected; "z" accepted
]

for word in test_words:
    result = is_valid_simplified_english_word(word)
    print(f"{word!r}: {result}")
```
